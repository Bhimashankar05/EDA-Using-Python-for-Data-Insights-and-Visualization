# Exploratory Data Analysis of Unicorn Companies Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Unicorn_Companies.csv')
df
```

```python
df.shape   # 1037 rows and 13 columns
```

```python
df.columns
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
del df['Financial Stage']
del df['Portfolio Exits']
```

```python
df = df.rename(columns={'Valuation ($B)': 'Valuation'})
df
```

```python
cols = ['Valuation']
for i in cols:
    df[i] = df[i].map(lambda x: str(x).lstrip('$')).astype(float)

df
```

# Top 10 companies as per top valuation

```python
data_company_valuation = df.loc[:,['Company','Valuation']].head(10)
data_company_valuation
```

```python
data_company_valuation.Company
```

```python
data_company_valuation.Valuation
```

```python
plt.bar(data_company_valuation.Company,data_company_valuation.Valuation,width = 0.5,color = 'y')
plt.xticks(rotation = 45)
plt.show()
```

# Conclusion1 :- Bytedance company has greatest valuation  and Revolute has lowest

# Finding out the largest % of area of Industry

```python
df.Industry
```

```python
df.Industry.value_counts().head()
```

```python
data = df.Industry.value_counts().head()
data
plt.pie(data,autopct = '%.f%%',labels = ['Fintech','Internet software & services','E-commerce & direct-to-consumer','Artificial intelligence','Health'],
                                   explode = [0.2,0,0,0,0])
plt.show()
```

# Conclusion 2:- 

The Fintech industrys are the topmost at various Unicorn comapnies which is there after followed by Internet software & services.The E-commerce industrys is having only 17% in companies and 11% same for both Health and AI industries.Other than these five all the remaining are lower % than this.
