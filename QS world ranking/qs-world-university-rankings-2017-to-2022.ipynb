# QS-world-university-rankings-2017-to-2022

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('qs-world-university-rankings-2017-to-2022-V2.csv')
df
```

```python
df.drop(labels=['link','logo','size'],axis = 1,inplace=True)
df
```

```python
df.rank_display = df.rank_display.fillna(0)
```

```python
df[df.rank_display.isna()].sum()
```

# lets see the percentwise private and public university 

```python
df.type.value_counts().plot(kind  = 'pie',autopct = '%.f%%',explode = [0,0.2])
plt.show
```

# Conclusion 1: only 18% are the private Universities remaing 82% are Public

```python
# lets see university typewise reasearch ratio
```

```python
sns.histplot(data = df,x = 'research_output',hue = 'type')
```

# Conclusion 2:- Near about 3200 Public Universities do the very high reasearch and almost 700 Public universities do High research whereas private universities are lagging so much in research and doing the medium range research

```python
# sort the yearwise top 10 rank universities with score
year_2017 = df[df['year'] == 2017].sort_values(by = 'score',axis = 0,ascending=False).head(10)
year_2017 
```

```python
year_2018 = df[df['year'] == 2018].sort_values(by = 'score',axis = 0,ascending=False).head(10)
year_2018 
```

```python
year_2019 = df[df['year'] == 2019].sort_values(by = 'score',axis = 0,ascending=False).head(10)
year_2019
```

```python
year_2020 = df[df['year'] == 2020].sort_values(by = 'score',axis = 0,ascending=False).head(10)
year_2020
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=1)
fig = plt.figure(figsize=(16, 6))
plt.tight_layout()

plt.subplot(1, 4, 1)
plt.bar(x = year_2017['university'] , height =year_2017['score'] ,width = 0.5)
plt.xticks(rotation = 90)
plt.ylim(90,100)
plt.title('Year 2017')

plt.subplot(1,4,2)
plt.bar(x = year_2018['university'] , height =year_2018['score'] ,width = 0.5)
plt.xticks(rotation = 90)
plt.ylim(90,100)
plt.title('Year 2018')

plt.subplot(1,4,3)
plt.bar(x = year_2019['university'] , height =year_2019['score'] ,width = 0.5)
plt.xticks(rotation = 90)
plt.ylim(90,100)
plt.title('Year 2019')

plt.subplot(1,4,4)
plt.bar(x = year_2020['university'] , height =year_2020['score'] ,width = 0.5)
plt.xticks(rotation = 90)
plt.ylim(90,100)
plt.title('Year 2020')


plt.xlabel('score')
plt.ylabel('Universities')

plt.show()
```

# Conclusion 3:- From above graphs it is clear that all the top ranked five universities are possessing same rank with almost same range score throughout 4 years.There is small variation in the scores and ranks of universities from 6 to 10.The almost range of  score of top 10 universities is in between 90 to 100. The MIT is top ranked university for all 4 years.

# lets try to find relation between student faculty ratio  and  rank

```python
year_2017[['rank_display','student_faculty_ratio','faculty_count']]
```

```python
plt.plot(year_2017['rank_display'],year_2017['student_faculty_ratio'],'o-g')
```

```python
plt.plot(year_2018['rank_display'],year_2018['student_faculty_ratio'],'o-g')
```

```python
plt.plot(year_2019['rank_display'],year_2019['student_faculty_ratio'],'o-g')
```

```python
plt.plot(year_2020['rank_display'],year_2020['student_faculty_ratio'],'o-g')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=1)
fig = plt.figure(figsize=(16, 6))
plt.tight_layout()

plt.plot(year_2017['rank_display'],year_2017['student_faculty_ratio'],'o-g')

plt.plot(year_2018['rank_display'],year_2018['student_faculty_ratio'],'D-r')

plt.plot(year_2019['rank_display'],year_2019['student_faculty_ratio'],'*-y')

plt.plot(year_2020['rank_display'],year_2020['student_faculty_ratio'],'^-k')

plt.legend(['Year 2017','Year 2018','Year 2019','Year 2020'])
plt.title('Relationship between QS rank of top 10 universities and Student Faculty Ratio ')
plt.xlabel('Rank')
plt.ylabel('Student Faculty ratio')
plt.show()
```

# Conclusion 5:- From above comparative study between student faculty ratio and rank, it is clear that in universities where student faculty ratio is minimum that means lesser students allotted to each faculty then it has good rank. But as student faculty ratio becomes poor then it effects on decreasing the ranking of that university line from 6 to 10 ranked universities.
